**Problem 3**: For $A:\mathbb{C}^{n}\rightarrow \mathbb{C}^{m}$, show that $\left<A{\bf x},{\bf y} \right> = \left<{\bf x}, A^{\dagger}{\bf y} \right>$, where $A^{\dagger}_{ij} = A^{\ast}_{ji}$ and
+
$$
\left(A{\bf x}\right)_{l} = \sum_{j=1}^{n}A_{lj}x_{j}, ~ l=1, \cdots, m.
$$


### Proof:


$$
\langle A{\bf x}, {\bf y} \rangle
= \sum_{l=1}^{m} (A{\bf x})_{l} \, y_{l}^{\ast}.
= \sum_{l=1}^{m} \left( \sum_{j=1}^{n} A_{lj} x_{j} \right) y_{l}^{\ast}.
= \sum_{j=1}^{n} x_{j} \left( \sum_{l=1}^{m} A_{lj} y_{l}^{\ast} \right).
$$

Now note that
$$
\sum_{\ell=1}^m A_{\ell j} y_\ell^\ast
= \left(\sum_{\ell=1}^m A^\ast_{\ell j} y_\ell\right)^\ast.
$$

But
$$
(A^\dagger \mathbf{y})_j = \sum_{\ell=1}^m A^\ast_{\ell j} y_\ell,
$$

so
$$
\sum_{\ell=1}^m A_{\ell j} y_\ell^\ast = \big( (A^\dagger \mathbf{y})_j \big)^\ast.
$$

Therefore,
$$
\langle A\mathbf{x}, \mathbf{y} \rangle
= \sum_{j=1}^n x_j \,(A^\dagger \mathbf{y})_j^\ast
= \langle \mathbf{x}, A^\dagger \mathbf{y} \rangle.
\qquad
$$


**Problem 4**: (Graduate Level/Extra Credit) In class, we showed for matrix $A$ such that $A:\mathbb{C}^{n}\rightarrow \mathbb{C}^{m}$ that

$$
\text{ker}(A^{\dagger}) \perp \text{rng}(A).
$$

Now show that $\mathbb{C}^{m} = \text{ker}(A^{\dagger}) \oplus \text{rng}(A)$.  To do this, suppose that $\exists {\bf y}\in \mathbb{C}^{m}, {\bf y}\neq 0, ~{\bf y} \notin \text{ker}(A^{\dagger}) \oplus \text{rng}(A)$.  
 * Why can we assume without loss of generality that ${\bf y}\perp \text{ker}(A^{\dagger})$ and ${\bf y}\perp\text{rng}(A)$?
 * Given the above, we have $\forall {\bf x}\in \mathbb{C}^{n}, \left<{\bf y}, A{\bf x}\right>=0$.  What choice of ${\bf x}$ shows us that ${\bf y}=0$?
 * Why does this give you the result?  


## Proof by contradiction

We stated that
$$
\ker(A^\dagger) \oplus \text{rng}(A) \subseteq \mathbb{C}^m
$$
$$
\text{now suppose } \exists
\vec{\omega} \in \mathbb{C}^m, \quad \vec{\omega} \neq 0, \text{ s.t. }
\vec{\omega} \perp \ker(A^\dagger) \text{ and } \vec{\omega} \perp \text{rng}(A), \text{ because of Gram Schmidt }
$$
$$
\text{If } \vec{\omega} \perp \text{rng}(A), \forall \vec{x} \in \mathbb{C}^n,
\langle \vec{\omega}, A\vec{x} \rangle = \langle A^\dagger \vec{\omega}, \vec{x} \rangle =0.
$$
Let
$$
\vec{x} = A^\dagger \vec{\omega} \in \mathbb{C}^n.
$$

$$
 \langle A^\dagger \vec{\omega}, A^\dagger \vec{\omega} \rangle
= 0 \quad \forall \vec{x} \in \mathbb{C}^n, \text { which implies }
A^\dagger \vec{\omega} = 0,
$$

$$
\vec{\omega} \in \ker(A^\dagger), \text{ but }\vec{\omega} \perp \ker(A^\dagger), \text{ making }
\langle \vec{\omega}, \vec{\omega} \rangle = 0,
$$
which means $\vec{\omega} = 0$.

but we said $\vec{\omega} \neq 0$.

Therefore no such $\vec{\omega}$ exists, and we conclude:

$$
\mathbb{C}^m = \ker(A^\dagger) \oplus \text{rng}(A).
$$

Let
$$
\operatorname{rng}(A) = \mathrm{span}\{e_j\}_{j=1}^r, \quad \{e_j\} \ \text{ are all linearly independent}
$$
Then $e_j = A x_j$ for some $x_j$, hence
$$
A^\dagger e_j = A^\dagger A x_j.
$$

Suppose
$$
\sum_{j=1}^r c_j A^\dagger e_j = 0.
$$
For all $x$,
$$
0 = \Big\langle \sum_{j=1}^r c_j A^\dagger e_j, x \Big\rangle
= \sum_{j=1}^r c_j \langle e_j, A x \rangle.
$$
Since $\{e_j\}$ is linearly independent in $\operatorname{rng}(A)$, all $c_j=0$.  

Thus $\{A^\dagger e_j\}_{j=1}^r$ is linearly independent, so
$$
\dim(\operatorname{rng}(A^\dagger)) \geq r = \dim(\operatorname{rng}(A)).
$$

Together:
$$
\dim(\operatorname{rng}(A^\dagger)) = \dim(\operatorname{rng}(A)).
$$
