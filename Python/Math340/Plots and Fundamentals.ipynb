```python
x = np.linspace(-np.pi, 2*np.pi, 1000)
plt.plot(x, np.sin(x), color = 'blue')
plt.xlabel('x')
plt.ylabel('$\sin(x)$')
plt.title('Sine Wave')
plt.show()
plt.clf()
```

```python
def square(x):
    return x*x
plt.plot(np.linspace(-3,2,100), square(np.linspace(-3,2,100)), color = 'red')
plt.xlabel('x')
plt.ylabel('x^2')
plt.title('Square Wave')
plt.show()
plt.clf()
```

```python
from mpl_toolkits import mplot3d
def f(x,y):
  return 1 - x**2 - y**2
x = y = np.linspace(-1,1,100)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
fig = plt.figure(figsize = (16,12))
ax = plt.axes(projection = '3d')
ax.contour3D(X, Y, Z, 200, cmap = 'binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
