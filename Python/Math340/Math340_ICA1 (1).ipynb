```python
import math
import numpy as np
import matplotlib.pyplot as plt
sigma = 40
dt = .1
amp = 1.0
n_values = []
r1_values = []
for n in range(1,30):
    r1 = np.float64(amp)
    arg = sigma*dt
    for i in range(n):
        r1 = np.float64(r1*math.exp(-arg))
    for i in range(n):
        r1 = np.float64(r1*math.exp(arg))
    n_values.append(n)
    r1_values.append(r1)
    print(' %d times exp(-sigma dt) anf exp (sigma dt): %.16f'%(n,r1))
    
plt.plot(n_values, r1_values, marker='o', linestyle='-')
plt.xlabel("n (Iterations)")
plt.ylabel("r1 Value")
plt.title("Effect of Repeated exp(-sigma*dt) and exp(sigma*dt)")
plt.grid()
plt.show()
    
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt
sigma = 40
dt = .1
amp = 1.0
n_values = []
r1_values = []
for n in range(1,30):
    r1 = np.float32(amp)
    arg = sigma*dt
    for i in range(n):
        r1 = np.float32(r1*math.exp(-arg))
    for i in range(n):
        r1 = np.float32(r1*math.exp(arg))
    n_values.append(n)
    r1_values.append(r1)
    print(' %d times exp(-sigma dt) anf exp (sigma dt): %.16f'%(n,r1))
    
plt.plot(n_values, r1_values, marker='o', linestyle='-')
plt.xlabel("n (Iterations)")
plt.ylabel("r1 Value")
plt.title("Effect of Repeated exp(-sigma*dt) and exp(sigma*dt)")
plt.grid()
plt.show()
    
```

Errors grow as n increases when working with exponents. In float64, we see the percision start to drop off at n = 187. from n = 179 to 186, the values start to oscillate from right below 1 to right above 1 then the rest of the values from n = 187 to n = 200, the output is 0. In float32, we see a similar pattern, but starts much quicker. from n = 1 to n = 21, the out is 1 with 16 zeros after the decimal. from n = 22 to n = 25, we see a little oscillation from below 1 to above 1. Then at n = 26 to n = 200, the output is 0 because we have meet our floating point percison thershold. The difference between float64 and float32 is hoq quickly percision is lost. The code only has us printing out 16 digits so we are able to see how much is lost within the first 16 digits. In general, floating point percision is proportional to $2^{(n-1)} - 1$. For small n, error is tiny. Large n, error can be great. float32 is single precision which has 32 total bits (~ 7 decimal places). 1 bit for the sign, 8 bits for the exponent, and 23 bits for mantissa (effectivly 24). 58 iterations is the threshold for 24 bits. float64 is double precision which has 64 bits (~15-16 decimal places). 1 bit for the sign, 11 bits for the exponent, 52 bits for mantissa (effectivly 53). With 53 bits of precision, float64 can handle significantly more accurate calculations before rounding errors accumulate. This is why the float32 experiment showed noticeable errors around 58 iterations, while float64 can go much further before numerical precision is significantly affected

```python

import cv2
from paddleocr import PaddleOCR
import subprocess
import os

# Set the path to the image you want to process
image_path = r"C:\Users\megan\LinAlg\Math 340\ICA_Image.jpg"

# Load the image using OpenCV
image = cv2.imread(image_path)

# Convert the image to grayscale (useful for OCR)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Initialize the PaddleOCR tool (for better text extraction of math equations)
ocr = PaddleOCR(use_angle_cls=True, lang="en")

# Perform OCR on the image to extract text
result = ocr.ocr(image_path, cls=True)

# Initialize an empty string to store the LaTeX formatted equations
latex_text = ""

# Extract text from OCR result and format it for LaTeX
for line in result:
    for word in line:
        latex_text += word[1][0] + "\n"  # Append the recognized text (equations)

# Print the LaTeX text to see what was extracted
print(latex_text)

# Specify the path for the LaTeX file
latex_file_path = "ICA_Image.tex"

# Create and write LaTeX formatted content into the .tex file
with open(latex_file_path, "w") as f:
    f.write("\\documentclass{article}\n\\usepackage{amsmath}\n\\begin{document}\n")
    f.write("$$\n" + latex_text + "\n$$\n")  # Formatting for LaTeX math mode
    f.write("\\end{document}")

print(f"LaTeX file saved as {latex_file_path}")

# Run pdflatex to convert the .tex file into a PDF
try:
    subprocess.run(["pdflatex", latex_file_path], check=True)
    print("PDF generated successfully!")
except subprocess.CalledProcessError as e:
    print("Error generating PDF:", e)
```
