Sigmoid Function solutions to $U' = \sigma U(1-U)$:
$$U = \frac{u_{0}e^{\sigma t}}{1-u_{0}+u_{0}e^{\sigma t}}$$
$$V = \frac{v_{0}e^{\sigma t}}{1-v_{0}+v_{0}e^{\sigma t}}$$
With initial conditons:
$U(0) = u_{0} $
$V(0) = u_{0} + \epsilon$


## A1

Compute $\frac{1}{\Delta t} \ln{\frac{Z_{n+1}}{Z_{n}}}$ 

```python
import numpy as np
import matplotlib.pyplot as plt
o = 2
dt = 0.001
t = np.arange(0+dt,10+dt,dt)
u0 = 0.002
eps = .00001
v0 = u0 + eps
growth_rates = []
for i in t:
    num = (u0 * np.exp(o * (i + 1))) / (1 - u0 + u0 * np.exp(o * (i + 1)))
    denom = (u0 * np.exp(o * i)) / (1 - u0 + u0 * np.exp(o * i))
    growth_rates.append(np.log(num/denom))
plt.plot(t,np.array(growth_rates)/dt)
plt.title(r'$\frac{1}{\Delta t} \ln{\frac{Z_{n+1}}{Z_{n}}}$')
plt.grid()
plt.show()
```

## A2

compute $\sigma(1-U)$

```python
values1 = []
for i in t:
    values1.append(o*(1-(u0* np.exp(o*i)) / (1 - u0 + u0*np.exp(o*i))))
plt.plot(t,values1)
plt.title(r'$\sigma(1-U)$')
plt.grid()
plt.show()
```

## B1

compute $\frac{1}{\Delta t}\ln{\frac{U_{n+1}-V_{n+1}}{U_{n}-V_{n}}}$

```python
values2 = []
for i in t:
    num = (u0 * np.exp(o * (i + 1))) / (1 - u0 + u0 * np.exp(o * (i + 1))) -(v0 * np.exp(o * (i + 1))) / (1 - v0 + v0 * np.exp(o * (i + 1)))
    denom =  (u0 * np.exp(o * (i))) / (1 - u0 + u0 * np.exp(o * (i))) - (v0 * np.exp(o * (i))) / (1 - v0 + v0 * np.exp(o * (i)))
    values2.append(np.log(num/denom))
plt.plot(t,np.array(values2)/dt)
plt.title(r'$\frac{1}{\Delta t}\ln{\frac{U_{n+1}-V_{n+1}}{U_{n}-V_{n}}}$')
plt.grid()
plt.show()
```

## B2

compute $\sigma (1-2U)$

```python
values3 = []
for i in t:
    values3.append(o*(1-2*((u0* np.exp(o*i)) / (1 - u0 + u0*np.exp(o*i)))))
plt.plot(t,values3)
plt.title(r'$\sigma(1-2U)$')
plt.grid()
plt.show()
```
