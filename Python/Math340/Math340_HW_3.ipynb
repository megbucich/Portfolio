### Results for Q2

Create a Python function for each of the three activation functions: Logistic, Hyperbolic Tangent, and Rectified Linear Unit (ReLU), based on the descriptions provided in the Wikipedia article on activation functions

Logistic: $$\frac{1}{1 + e^{-x}}$$
Hyperbolic Tangent: $$\frac{e^x - e^{-x}}{e^x + e^{-x}}$$
Rectified Linear Unit: $$\text{max}(0,x)\text{ for } x > 0$$

```python
import numpy as np
import matplotlib.pyplot as plt
import math

def logistic(x):
    return( 1 / (1 + math.exp(-x)))

def hyperbolic_tangent(x):
    return( ((math.exp(x)) - (math.exp(-x))) / ((math.exp(x)) + (math.exp(-x))))
    
def ReLU(x):
    if x <= 0:
        return(0)
    if x > 0:
        return(x)
        
x = np.linspace(-10,10,2000)
y1 = []
y2 = []
y3 = []
for i in x:
    y1.append(logistic(i))
    y2.append(hyperbolic_tangent(i))
    y3.append(ReLU(i))

plt.figure(figsize = (12,4))
plt.subplot(1,3,1)
plt.plot(x,y1)
plt.grid()
plt.title("Logistic")

plt.subplot(1,3,2)
plt.plot(x,y2)
plt.grid()
plt.title("Hyperbolic Tangent")

plt.subplot(1,3,3)
plt.plot(x,y3)
plt.grid()
plt.title("Rectified Linear Unit (ReLU)")

plt.show()
```

```python
def left_eye(x):
    return(np.sqrt(1.5-(x-1.5)**2) + 1)
def right_eye(x):
    return(np.sqrt(1.5-(x+1.5)**2) + 1)
def mouth(x):
    return(-(np.sqrt(6-x**2)))
def outline1(x):
    return(np.sqrt(16 - x**2))
def outline2(x):
    return(-(np.sqrt(16 - x**2)))
x1 = np.linspace(.2753,2.7247,100)
x2 = np.linspace(-2.7247,-.2753,100)
x3 = np.linspace(-(np.sqrt(6)),np.sqrt(6),100)
x4 = np.linspace(-4,4,100)
y1 = []
for i in x1:
    y1.append(left_eye(i))
y2 = []
for i in x2:
    y2.append(right_eye(i))
y3 = []
for i in x3:
    y3.append(mouth(i))
y4 = []
for i in x4:
    y4.append(outline1(i))
y5 = []
for i in x4:
    y5.append(outline2(i))

plt.xlim(-10,10)
plt.ylim(-10,10)
plt.plot(x1,y1,'b')
plt.plot(x2,y2,'b')
plt.plot(x3,y3,'b')
plt.plot(x4,y4,'b')
plt.plot(x4,y5,'b')
plt.grid()
plt.show()




```

### Results for Q5

```python
import math
inc = 3
for n in range (1,60, inc):
    r = 2
    for i in range(n):
        r = math.sqrt(r)
    for i in range(n):
        r = r**2
    print('%d times sqrt and **2: %.16f'%(n,r)) #is this even english Lol
```

The way the code is set up is we start with n = 1, it goes through the loop, then n += 3 until we get to the end of the range, which is what "for n in range(1,60, inc)" is doing. As each n value goes through the loop, we have r = 2. Depending on what n value we are on, we take the square root of r, n times. Then, we square that value, n times. Essentially "undoing" what we did in the first for loop. I guess the point is that the larger n is, the further we move away from 2. When n = 1, we square root 2, then square it to get back to 2. But as we repeat this process as n increases, the answer becomes less and less accurate. Around n = 52, the precision drops off drastically from 1.86ish to 1.0. This must indicate the limit of percision.
