```python
import matplotlib.pyplot as plt
import numpy as np
def logistic(x):
    return( 1 / (1 + np.exp(-x)))
t = np.arange(-10,11,.01) #numpy array
x = logistic(t) #also an array
y = np.diff(x) 

plt.subplot(1,2,1)
plt.annotate('local max', xy = (5,1), xytext = (-8,1), arrowprops = dict(facecolor = 'b', shrink = 0.05))
plt.plot(t,x)
plt.grid()
plt.subplot(1,2,2)
plt.plot(t[1:],y) #cannot directly plot x agaisnt y so we just take the last 9 element of t and employ np.diff(x) on there were x is the logistic 
plt.annotate('local max', xy = (0,.0025), xytext = (-10,.0025), arrowprops = dict(facecolor = 'b', shrink = 0.05))
plt.grid()
plt.show()
```
