### Superimposed Plots M = 8

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
m = 8
n = 100
plt.figure(figsize=(10, 6))
norm_A = []
condition_numbers = []
sigma_mins = []
for i in range(n):
    A = np.random.randn(m, m) / np.sqrt(m)  
    evalues_A, _ = np.linalg.eig(A) 
    plt.scatter(evalues_A.real, evalues_A.imag, alpha=0.4, color="blue")
    norm_A.append(np.linalg.norm(A,2))
    singular_v = np.linalg.svd(A, compute_uv = False) #just want the singular values, not U and V
    sigma_max = singular_v.max()
    sigma_min = singular_v.min()
    sigma_mins.append(sigma_min)
    k = sigma_max / sigma_min if sigma_min > 0 else np.inf #wan't to avoid dividing by zero
    condition_numbers.append(k)
    

plt.title(f"Superimposed Eigenvalues of 100 Random Matrices (m={m})")
plt.xlabel("Real Part")
plt.ylabel("Imaginary Part")
plt.grid()
plt.show()

plt.scatter(range(n),sigma_mins, color = "pink")
plt.axhline(y=2**(-1), color='red', linestyle='--', label=r"$2^{-1}$")
plt.axhline(y=4**(-1), color='green', linestyle='--', label=r"$4^{-1}$")
plt.axhline(y=8**(-1), color='blue', linestyle='--', label=r"$8^{-1}$")
plt.title(r"$\sigma_{{min}}$ of 100 Random Matrices (m={})".format(m))
plt.xlabel("Random Matrix")
plt.ylabel(r"$\sigma_{min}$")
plt.legend()
plt.grid()
plt.show()
print(f"The largest eigenvalue is \033[43m{max(np.abs(evalues_A))}\033[0m") 
print(f"The average 2-norm is  \033[45m{np.mean(norm_A)}\033[0m")
print(f"The average conditon number is \033[46m{np.mean(condition_numbers)}\033[0m")

cutoff =[2**-1,4**-1,8**-1]
proportions = [np.mean(np.array(sigma_mins) < c) for c in cutoff]
for i in range(len(cutoff)):
    print(f"Proportion of matrices with σ_min < {cutoff[i]:.3f}: {proportions[i]:.2%}")
```

### What does the tail of the probability distribution of smallest singular values look like? How does the scale of all this change with $m$?

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
matrix_sizes = [8, 16, 32, 64, 128]
n = 100
plt.figure(figsize=(12, 8))
for m in matrix_sizes:
    sigma_mins = []
    for i in range(n):
        A = np.random.randn(m, m) / np.sqrt(m)  
        singular_v = np.linalg.svd(A, compute_uv=False)
        sigma_min = singular_v.min()
        sigma_mins.append(sigma_min)
    sns.kdeplot(sigma_mins, label=f"m={m}", fill=True)
    p5 = np.percentile(sigma_mins,5) #look at the 5th percentile (tail)
    print(f"m={m}: 5th percentile of σ_min = {p5:.4f}")
plt.title(r"Distribution of $\sigma_{min}$ for 100 Random Matrices")
plt.xlabel(r"$\sigma_{min}$")
plt.legend(title="Matrix Size (m)")
plt.grid()
plt.show()

```
